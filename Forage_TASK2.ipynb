**REACTION TYPES DATAFRAME**

```python
rtype=pd.read_csv('ReactionTypes.csv',index_col=0)
rtype
```

```python
rtype.rename(columns={'Type':'Reaction Type'},inplace=True)
```

```python
rtype
```

```python
rtype.dtypes
```

```python
reaction=pd.read_csv('Reactions.csv',index_col=0)
reaction
```

```python
reaction=reaction.drop(columns='User ID')
```

```python
reaction.isna().sum()
```

```python
reaction.dropna(subset='Type',inplace=True)
```

```python
reaction.isna().sum()
```

```python
reaction.dtypes
```

```python
reaction.rename(columns={'Type':'Reaction Type'},inplace=True)
```

```python
reaction
```

**MERGING ALL THE DATAFRAMES WITH REACTION AS BASE DATAFRAME**

```python
final_df=reaction.merge(content,on='Content ID',how='left')
final_df=final_df.merge(rtype,on='Reaction Type',how='left')
```

```python
final_df
```

```python
final_df.isna().sum()
```

**FINDING TOP 5 CATEGORIES**

```python
totalscore = final_df.groupby('Category')['Score'].sum()
totalscore
```

```python
top5=totalscore.nlargest(5)
print(top5,'\n')
top5_category=top5.index
```

```python
result=final_df[final_df['Category'].isin(top5_category)]
result
```

```python
result.isna().sum()
```

```python
result.to_csv('End Result')
```
